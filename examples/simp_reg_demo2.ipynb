## Regression Demo: comparing models with multi mripls

We use two mripls to compare a quadratic and fourier model on the same dataset. 

```python
# terminal: run 'ipcluster start --n=<no_engines>
from venture.venturemagics.ip_parallel import *; lite=False; clear_all_engines()


quadratic='''
[assume w0 (normal 0 3) ]
[assume w1 (normal 0 3) ]
[assume w2 (normal 0 3) ]
[assume x (mem (lambda (i) (normal 0. 5.) ) )]
[assume noise (gamma 2 1) ]
[assume f (lambda (x) (+ w0 (* w1 x) (* w2 (* x x)) ) ) ]
[assume y (mem (lambda (i) (normal (f (x i) ) noise) ) )]
[assume n (gamma 1 1)]
'''
fourier='''
[assume w0 (normal 0 3) ]
[assume w1 (normal 0 3) ]
[assume omega (normal 0 3) ]
[assume theta (normal 0 3) ]
[assume x (mem (lambda (i) (normal 0. 5.) ) )]
[assume noise (gamma 2 1) ]
[assume f (lambda (x) (+ w0 (* w1 (sin (+ (* omega x) theta) ) ) ) ) ]
[assume y (mem (lambda (i) (normal (f (x i) ) noise) ) )]
[assume n (gamma 1 1)]
'''

vq = MRipl(4,lite=lite,verbose=False); vq.execute_program(quadratic)
vf = MRipl(4,lite=lite,verbose=False); vf.execute_program(fourier)
vs=[vq,vf]

# function for generating data
def generate_sin_data(n):
    xs = np.random.normal(loc=0,scale=2.5,size=n)
    ys = 3*np.sin(xs)
    xys = zip(xs,ys)
    fig,ax = subplots(figsize=(6,2))
    ax.scatter(xs,ys); ax.set_title('Data for regression problem')
    return xys

n=6
xys = generate_sin_data(n)

# load data as observation for each mripl and do some inference
def mripls_infer(vs,xys,no_transitions,n):
    for i,(x,y) in enumerate(xys):
        [v.observe('(x %i)' % i , '%f' % x) for v in vs]
        [v.observe('(y %i)' % i , '%f' % y) for v in vs]
    [v.observe('n','%i'%n) for v in vs]
    [v.infer(no_transitions) for v in vs];
    
mripls_infer(vs,xys,200,n)
```

We now plot the curves f inferred from the two models of the data. 

```python
%%mr_map vq plot_curves store_plot_q
import numpy as np

def plot_curves(ripl):
    n = int( np.round( ripl.sample('n') ) )
    xs = [ripl.sample('(x %i)' % i) for i in range(n)]
    ys = [ripl.sample('(y %i)' % i) for i in range(n)]
    xr = np.linspace(1.5*min(xs),1.5*max(xs),100)
    f_xr = [ripl.sample('(f %f)' % x) for x in xr]
    
    noise = ripl.sample('noise')
    f_a = [fx+noise for fx in f_xr]
    f_b = [fx-noise for fx in f_xr]
    
    fig,ax = plt.subplots(figsize=(6,2))
    ax.scatter(xs,ys)
    ax.set_color_cycle(['m', 'gray','gray'])
    ax.plot(xr,f_xr,xr,f_a,xr,f_b)
    ax.set_title('Observed x,y data plotted with inferred function f (+- 1sd of noise)')
    
    return None
```

### More data, more ripls

We now consider a larger number of ripls and data. Now the fourier model will perform better on average. 

```python
## Posterior samples from the models, with many ripls and larger amount of data

no_ripls = 10
vq = MRipl(no_ripls,lite=lite,verbose=False); vq.execute_program(quadratic)
vf = MRipl(no_ripls,lite=lite,verbose=False); vf.execute_program(fourier)
vs=[vq,vf]
no_transitions=500
n=20
xys = generate_sin_data(n); xs = [pair[0] for pair in xys]
xr = np.linspace(1.3*min(xs),1.3*max(xs),30)

mripls_infer(vs,xys,no_transitions,n);
```

```python
## samples from the posterior from each model (with the actual generative model plotted also)
y_vals = [ [v.sample('(normal (f %f) noise)' % x) for x in xr] for v in vs]
fig,ax = subplots(1,2, figsize=(13,6) )
for i in range(2):
    [ax[i].scatter(xr,[ x[ripl] for x in y_vals[i]],s=6) for ripl in range(no_ripls) ]
    ax[i].plot(xr,3*np.sin(xr))
ax[0].set_title('Posterior samples for quadratic model')
ax[1].set_title('Posterior samples for fourier model');
```

```python
# Comparing model performance using the mean logscore
mean_logscores = [ np.mean( v.get_global_logscore() ) for v in vs ]
print 'logscores. [quadratic,fourier] = ',mean_logscores
```

### Combining the models

We now form a mixture of both quadratic and fourier models and apply it to the dataset above. From the previous cells, we know that the fourier model performs better on average and that it contains the true generative function. So we expect that with enough inference the mixture model will become close in logscore to the fourier model and will be bounded above by it. 

```python
quad_fourier='''
[assume w0 (normal 0 3) ]
[assume w1 (normal 0 3) ]
[assume w2 (normal 0 3) ]
[assume omega (normal 0 3) ]
[assume theta (normal 0 3) ]
[assume x (mem (lambda (i) (normal 0 5) ) )]
[assume noise (gamma 2 1) ]

[assume model (if (flip) 1 0) ]
[assume quadratic (lambda (x) (+ w0 (* w1 x) (* w2 (* x x) ) ) ) ]
[assume fourier (lambda (x) (+ w0 (* w1 (sin (+ (* omega x) theta) ) ) ) ) ]
[assume f (if (= model 0) quadratic fourier) ]

[assume y (mem (lambda (i) (normal (f (x i) ) noise) ) )]
[assume n (gamma 1 1)]
'''
no_ripls = 20
vqf = MRipl(no_ripls,lite=lite,verbose=False); vqf.execute_program(quad_fourier)
vs=[vqf]
mripls_infer(vs,xys,no_transitions,n);   
```

```python
## logscore of mixture model
print 'logscores. [quadratic,fourier] = ',mean_logscores
print 'mixture model:', np.mean(vqf.get_global_logscore())
```

```python
# Posterior sampels for mixture model
y_vals = [vqf.sample('(normal (f %f) noise)' % x) for x in xr]
fig,ax = subplots(1,1, figsize=(10,4) )
[ax.scatter(xr,[ x[ripl] for x in y_vals],s=6) for ripl in range(no_ripls) ]
ax.plot(xr,3*np.sin(xr))
```
